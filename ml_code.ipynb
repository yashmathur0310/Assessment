## Loading the libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
pd.set_option('display.max_columns',None)
```

#### Reading the train dataset

```python
df_train=pd.read_csv('train.csv')
df_train
```

#### Checking for null values

```python
null=df_train.isnull().sum().sort_values(ascending=False)
```

```python
df_null=pd.DataFrame()
```

```python
df_null['columns']=null.index
df_null['values']=null.values
```

```python
df_null[:20]
```

```python
df_train.shape
```

##### Dropping the columns which has more than 40 percent null values

```python
df_train=df_train.drop(['Id','PoolQC','MiscFeature','Alley','Fence','FireplaceQu'],axis=1)
```

```python
df_train.isnull().sum()
```

```python
for col in df_train.columns:
    if df_train[col].isnull().sum()>0:
        if df_train[col].dtypes in ['int64','float64']:
            df_train[col]=df_train.fillna(df_train[col].mean())
        else:
            df_train[col]=df_train.fillna(df_train[col].mode()[0])
```

```python
df_train.isnull().sum()
```

###### Appending the categorical colums

```python
categorical=[]
for col in df_train.columns:
    if df_train[col].dtypes not in ['int64','float64']:
        categorical.append(col)
    else:
        continue
```

```python
len(categorical)
```

### Label encoding the categorical columns

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for col in categorical:
    df_train[col]=le.fit_transform(df_train[col])
```

```python
df_train
```

```python
x=df_train.drop(['SalePrice'],axis=1).values
y=df_train['SalePrice'].values
```

```python
print(x.shape)
print(y.shape)
```

## Taking 20 percent dataset for testing purpose

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
linear.fit(x_train,y_train)
```

```python
linear_pred=linear.predict(x_test)
```

```python
from sklearn.metrics import *
```

```python
print(mean_absolute_error(y_test,linear_pred))
print(mean_squared_error(y_test,linear_pred))
```

## Xgboost

```python
import xgboost as xgb
```

```python
xgboost=xgb.XGBRegressor()
xgboost.fit(x_train,y_train)
```

```python
pred_xgboost=xgboost.predict(x_test)
```

```python
print(mean_absolute_error(y_test,pred_xgboost))
print(mean_squared_error(y_test,pred_xgboost))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
```

```python
pred_rfr=rfr.predict(x_test)
```

```python
print(mean_absolute_error(y_test,pred_rfr))
print(mean_squared_error(y_test,pred_rfr))
```
